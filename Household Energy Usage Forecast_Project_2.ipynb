```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# load data in csv file using pandas 

data=pd.read_csv('household_power_consumption.csv')
data
```

```python
#check the unique for Sub_metering_3 feature columns

data['Sub_metering_3'].unique()
```

```python
#using for loop data type change for columns 

a=['Global_active_power','Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']
for i in a:
    data[i] = pd.to_numeric(data[i], errors='coerce')
```

```python
#data type change after check the information data 

data.info()
```

```python
#data type change for Date Column

data['Date'] = pd.to_datetime(data['Date'],dayfirst=True)
```

```python
data.info()
```

```python
#defult first 5 rows 

data.head()
```

```python
# Convert Time column to a datetime object
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M:%S', errors='coerce')
```

```python
data.info()
```

```python
#fill the Null value using mean()

data['Sub_metering_3'].fillna(data['Sub_metering_3'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
#using for loop for fill the Null value in dataset

b=['Global_active_power','Global_reactive_power','Voltage',	'Global_intensity',	'Sub_metering_1','Sub_metering_2']
for i in b:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
#afer check null value 

data.isnull().sum()
```

```python
#round of float data type columns 

data=data.round(2)
```

```python
#check the duplicates value in dataset 

data.duplicated().sum()
```

```python
data
```

```python
#Exract the Year,Month and Day

data['Year']=data['Date'].dt.year
data['Month']=data['Date'].dt.month
data['Day']=data['Date'].dt.day
```

```python
#Extract the Hour and Minute

data['Hour']=data['Time'].dt.hour
data['Minute']=data['Time'].dt.minute
```

```python
data1=data.drop(['Date','Time','DateTime'],axis=1)
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1.to_csv('Household Energy_Project_2.csv',index=False)
```

```python
datas=pd.read_csv('Household Energy_Project_2.csv')
datas.head()
```

```python
#Analysis for Numerical columns using Describe 

datas.describe()
```

```python
# Create rolling averages for smoothing

datas['Rolling_Avg_Power'] = datas['Global_active_power'].rolling(window=10).mean()
```

```python
#check the unique values in Rolling Average columns 

datas['Rolling_Avg_Power'].unique()
```

```python
#missing value apply mean value

datas['Rolling_Avg_Power'].fillna(datas['Rolling_Avg_Power'].mean(),inplace=True)
```

```python
#Missing values 
datas.isnull().sum()
```

```python
#round of data floating points 2 decimal 

house=datas.round(2)
```
