```python
data=pd.read_csv('project_6.csv')
```

```python
data
```

```python
#Splting the Target column and feature columns

x=data.drop(columns='Global_active_power',axis=1)
y=data['Global_active_power']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Apply the StanderScaler Method using Performance 

scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)
```

```python
#spliting Process for Training and Testing 

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Apply the Linear Regression Model

liner=LinearRegression()
liner.fit(x_train,y_train)
```

```python
#prediction

y_pred=liner.predict(x_test)
```

```python
#Training Accuracy Score for Linear Regression 

print('Linear Regression Model Training Accuracy Score:',liner.score(x_train,y_train))
```

```python
#Testing Accuracy score for Linear Regression 

print('Linear Regression Model Testing Accuracy Score:',liner.score(x_test,y_test))
```

```python
#Actual value and Machine Predicted Value

reg_model_diff = pd.DataFrame({'Actual Value' : y_test, 'Predicted Value' : y_pred})
reg_model_diff
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn import metrics
```

```python
#Evaluvate Matrics for Regression 

mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#Apply the Random Forest model

random=RandomForestRegressor()
random.fit(x_train,y_train)
```

```python
ran_y_pre=random.predict(x_test)
```

```python
#Training Accuracy score for Random Forest  

print('Random Forest Model Training Accuracy Score:',random.score(x_train,y_train))
```

```python
#Testing Accuracy score for Random Forest  

print('Random Forest Model Testing Accuracy Score:',random.score(x_test,y_test))
```

```python
#Evaluvate Matrics for Regression 

mae_ran = metrics.mean_absolute_error(y_test, ran_y_pre)
mse_ran = metrics.mean_squared_error(y_test, ran_y_pre)
r2_ran = np.sqrt(metrics.mean_squared_error(y_test, ran_y_pre))

print('Mean Absolute Error:', mae_ran)
print('Mean Square Error:', mse_ran)
print('Root Mean Square Error:', r2_ran)
```

```python
#Actual value and Machine Predicted Value using Random Forest Model

random_model_diff = pd.DataFrame({'Actual Value' : y_test, 'Predicted Value' : ran_y_pre})
random_model_diff
```
